```
#convert image to jpg
import cv2, os
base_path = "/content/drive/My Drive/final_year_project/train/ubiris2_1/CLASSES_400_300_Part1/"
new_path = "/content/drive/My Drive/final_year_project/train_jpg/train_image/"
for infile in os.listdir(base_path):
    print ("file : " + infile)
    read = cv2.imread(base_path + infile)
    outfile = infile.split('.')[0] + '.jpg'
    cv2.imwrite(new_path+outfile,read,[int(cv2.IMWRITE_JPEG_QUALITY), 200])
```

```
#convert image to mask
import cv2, os
base_path = "/content/drive/My Drive/final_year_project/train/ubiris2_1/ubiris/"
new_path = "/content/drive/My Drive/final_year_project/train_jpg/train_mask/"
for infile in os.listdir(base_path):
    print ("file : " + infile)
    read = cv2.imread(base_path + infile)
    outfile = infile.split('.')[0] + '.png'
    cv2.imwrite(new_path+outfile,read,[int(cv2.IMWRITE_JPEG_QUALITY), 200])
```
